```python
import torch
import torch.nn as nn
import numpy as np
```

```python
class AE(nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(784,256),
            nn.ReLU(),
            nn.Linear(256, 64),
            nn.ReLU(),
            nn.Linear(64,20),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(20, 64),
            nn.ReLU(),

            nn.Linear(64, 256),
            nn.ReLU(),

            nn.Linear(256, 784),
            nn.Sigmoid()
        )
    def forward(self, x):
        batchsz = x.size(0)
        x = x.view(batchsz, -1)
        x = self.encoder(x)

        x = self.decoder(x)

        x = x.view(batchsz, 1, 28, 28)

        return x
```

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

mnist_train = datasets.MNIST('mnist_data', train=True, transform=transforms.Compose([
    transforms.ToTensor()
]), download=True)

mnist_train = DataLoader(mnist_train, batch_size=32, shuffle=True)

mnist_test = datasets.MNIST('mnist_data', train=False, transform=transforms.Compose([
    transforms.ToTensor()
]), download=True)
mnist_test = DataLoader(mnist_test, batch_size=32, shuffle=True)
```

```python
mnist_train.dataset[0][0].shape
```

```python
epochs = 1000
lr = 1e-3
model = AE().to('cuda')
criteon = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
print(model)
```

```python
from visdom import Visdom
viz = Visdom()
for epoch in range(epochs):
    for batchidx, (x, _) in enumerate(mnist_train):
        x = x.cuda()
        x_hat = model(x)
        loss = criteon(x_hat, x)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    if epoch % 10 == 0:
        print('epoch:', epoch, 'loss:', loss.item())
    x, _ = mnist_test.__iter__().__next__()
    x = x.cuda()
    with torch.no_grad():
        x_hat = model(x)
    viz.images(x, nrow=8, win='x', opts=dict(title='x'))
    viz.images(x_hat, nrow=8, win='x_hat', opts=dict(title='x_hat'))
```
